- `pivot_longer()` "lengthens" data, increasing the number of rows and decreasing the number of columns. 

- `pivot_longer()` is an updated approach to `gather()`, designed to be both simpler to use and to handle more use cases.`gather()` isn't going away but is no longer under active development.
