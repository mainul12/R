# way 2

```r
df3 <- within(mpg, class <- factor(class, levels=names(sort(table(class), decreasing=TRUE))))
## plot
ggplot(df3,aes(x=class))+geom_bar()
```

```r
df3 <- within(mpg, class <- factor(class, levels=names(sort(table(class), decreasing=FALSE))))
## plot
ggplot(df3,aes(x=class))+geom_bar()
```

# Solution 3

```r
mpg %>%
    count(class) %>%
    mutate(class = fct_reorder(class, n, .desc = FALSE)) %>%
    ggplot(aes(x = class, y = n)) + geom_bar(stat = 'identity')
```
