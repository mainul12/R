```r
library(ggplot2)
```

```r
a <- rnorm(1000)
```

```r
df <- data.frame(a)
```

```r
head(a)
```

```r
class(df)
```

```r
ggplot(df, aes(x=a)) + 
    geom_histogram(bins = 100, color="black", fill="orange")
```

```r
some_data = rnorm(1000, mean = 90, sd = 10)
```

```r
some_data = data.frame(some_data)
```

```r
ggplot(some_data, aes(x=some_data))+
    geom_histogram(bins=100, color="black", fill="orange")
```
