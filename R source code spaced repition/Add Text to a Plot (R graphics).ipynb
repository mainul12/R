```r
plot(1:10, (1:10)**2, type = "p", xlab = "X", ylab = "Y")

text(x = 9, y = 81, labels = "(9, 81)")
text(x = 9, y = 81, labels = "(9, 81)", col = "orange", , adj = c(0.5, -1), cex = 1.5) ## -1 lifts the text above
text(x = 9, y = 81, labels = "(9, 81)", col = "blue", , adj = c(0, 2)) 
text(x = 9, y = 81, labels = "(9, 81)", col = "red", , adj = c(1, 3)) 
text(x = 9, y = 81, labels = "(9, 81)", col = "forest green", , adj = c(3, -1), cex = 2)
```
