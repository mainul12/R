Testing what `mtext` does

-> consult the `par omi` jupyter notebook

```r
options(repr.plot.width = 5, repr.plot.height = 5, repr.plot.res = 100)
```

```r
## creating a plot
plot(0:10, 0:10, type="s", xlab="X", ylab="Y") # type="n" hides the points
mtext("Testing what this does", side=3, line=0, adj=1.0, cex=1, col="forestgreen")  
## side argument says which side of the plot
## adding a text to the bottom
mtext("Adding a text to the bottom", side=1, line=0, adj=1.0, cex=1, col="orange")
mtext("Adding text to the left", side = 2, line = 0, adj = 1.0, cex = 1, col = "red")
```

```r
plot(1:10, 2*(1:10), xlab = "X", ylab = "Y")
mtext("Adding some text to the top", side = 3, col = "red")
mtext("Adding text to the bottom", side = 1, col = "blue")
mtext("Adding text to the left", side = 2, col = "purple")
mtext("Adding text to the right", side = 4, col = "orange")
```

```r
plot(1:10, 2*(1:10), xlab = "X", ylab = "Y")
mtext("Adding some text to the top", side = 3, line = 3, col = "red")
mtext("Adding text to the bottom", side = 1, line = 1, col = "blue")
mtext("Adding text to the left", side = 2, line = 1, col = "purple")
mtext("Adding text to the right", side = 4, line = 1, col = "orange")
```

```r
plot(1:10, (-4:5)^2, main = "Parabola Points", xlab = "xlab")
mtext("10 of them")
for(s in 1:4)
    mtext(paste("mtext(..., line= -1, {side, col, font} = ", s,
          ", cex = ", (1+s)/2, ")"), line = -1,
          side = s, col = s, font = s, cex = (1+s)/2)
mtext("mtext(..., line= -2)", line = -2)
mtext("mtext(..., line= -2, adj = 0)", line = -2, adj = 0)
##--- log axis :
plot(1:10, exp(1:10), log = "y", main = "log =\"y\"", xlab = "xlab")
for(s in 1:4) mtext(paste("mtext(...,side=", s ,")"), side = s)
```
