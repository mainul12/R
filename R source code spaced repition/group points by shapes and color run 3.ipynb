```r
library(wakefield)
library(tidyverse)
```

```r
options(repr.plot.width = 10, repr.plot.height = 6, repr.plot.res = 200)
```

```r
age <- as.integer(age(100))
```

```r
ggplot(NULL, aes(x = age)) + 
    geom_histogram(bins = 40)
```

```r
weight <- rnorm(100, 50, 3)
```

```r
category <- r_sample_factor(100, x = c("A", "B", "C"))
```

```r
df <- data.frame(age = age, weight = weight, category = category)
```

```r
head(df)
```

```r
ggplot(df, aes(x = age, y = weight, color = category)) + 
    geom_point()
```

```r
age1 <- as.integer(age(1000))
weight1 <- rnorm(1000, 50, 3)
category1 <- r_sample_factor(1000, x = c("A", "B", "C"))
df1 <- data.frame(age = age1, weight = weight1, category = category1)
ggplot(df1, aes(x = age, y = weight, color = category)) + 
    geom_point()
```

```r
table(df1$category)
```

```r
df1 %>% 
    group_by(category) %>%
    summarise(n = n())
```

```r
## changing the weight to rbinom

age1 <- as.integer(age(1000))
weight1 <- rbinom(1000, 10, 0.5)
category1 <- r_sample_factor(1000, x = c("A", "B", "C"))
df1 <- data.frame(age = age1, weight = weight1, category = category1)
ggplot(df1, aes(x = age, y = weight, color = category)) + 
    geom_point()
```
