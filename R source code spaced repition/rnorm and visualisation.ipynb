```r
library(ggplot2)
```

```r
options(repr.plot.width = 15, repr.plot.height = 9, repr.plot.res = 100)
```

```r
some_data <- data.frame(observations = rnorm(n=10000, mean = 0, sd = 1),
                        type = "A")
```

```r
head(some_data)
```

```r
ggplot(some_data, aes(x=observations)) +
  geom_histogram(bins=100, color="black",
                 fill= 'orange')
```

```r
plot_normal <- function(sample_size, mean, sd){
    some_data <- data.frame(observations = rnorm(n=sample_size, mean = mean, sd = sd),
                        type = "A")
    
    ggplot(some_data, aes(x=observations)) +
  geom_histogram(bins=100, color="black",
                 fill= 'orange') + 
    theme(text = element_text(size = 24)) 
}
```

```r
plot_normal(10000, mean = 10, sd = 100)
```
