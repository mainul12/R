type argument to plot: 

1-character string giving the type of plot desired. The following values are possible: 

- `p` for points 
- "l" for lines 
- "b" for both points and lines 
- "c" for empty points joined by lines 
- "o" for overplotted points and lines 
- "s" and "S" for stair steps  
- "h" for histogram-like vertical lines 

- Finally, "n" does not produce any points or lines.
